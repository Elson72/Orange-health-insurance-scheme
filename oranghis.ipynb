```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Set Seaborn style
sns.set_theme(style="whitegrid")

# Data for the health facilities
facilities = ['Complete Care Clinic', 'General Hospital', 'PHC Arogbo']
male_counts = [219, 288, 19]
female_counts = [160, 195, 10]
over_18 = [235, 330, 17]
under_18 = [144, 153, 12]
dependants = [199, 196, 14]
principals = [180, 287, 15]

# Total data counts for each facility
total_counts = [sum(counts) for counts in zip(male_counts, female_counts, over_18, under_18, dependants, principals)]

# 1. Percentage distribution of health facilities in a pie chart
plt.figure(figsize=(8, 6))
colors = sns.color_palette("pastel")
plt.pie(total_counts, labels=facilities, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Percentage Distribution of Health Facilities', fontsize=14, fontweight='bold')
plt.axis('equal')  # Ensures pie is drawn as a circle
plt.show()

# 2. Total aggregate of metrics data for ESE Odo in a bar chart
total_aggregate = [sum(metric) for metric in zip(male_counts, female_counts, over_18, under_18, dependants, principals)]
metrics = ['Total Male', 'Total Female', 'Total >18yrs', 'Total <18yrs', 'Total Dependants', 'Total Principals']

plt.figure(figsize=(12, 6))
sns.barplot(x=metrics, y=total_aggregate, palette="viridis")
plt.title('Total Aggregate Metrics Data for ESE Odo', fontsize=14, fontweight='bold')
plt.ylabel('Counts', fontsize=12)
plt.xlabel('Metrics', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
for i, v in enumerate(total_aggregate):
    plt.text(i, v + 3, str(v), ha='center', fontweight='bold')  # Adding values above the bars
plt.tight_layout()  # Adjust layout for better readability
plt.show()

# 3. Comparison of the 3 facilities' metrics data in a grouped bar chart
x = np.arange(len(metrics))  # Label locations
bar_width = 0.2

# Set colors for different bars
colors = sns.color_palette("Set2", 3)

plt.figure(figsize=(12, 8))
plt.bar(x - bar_width, [male_counts[0], female_counts[0], over_18[0], under_18[0], dependants[0], principals[0]], 
        width=bar_width, label='Complete Care Clinic', color=colors[0])
plt.bar(x, [male_counts[1], female_counts[1], over_18[1], under_18[1], dependants[1], principals[1]], 
        width=bar_width, label='General Hospital', color=colors[1])
plt.bar(x + bar_width, [male_counts[2], female_counts[2], over_18[2], under_18[2], dependants[2], principals[2]], 
        width=bar_width, label='PHC Arogbo', color=colors[2])

plt.title('Comparison of Health Facilities Metrics Data', fontsize=14, fontweight='bold')
plt.xlabel('Metrics', fontsize=12)
plt.ylabel('Counts', fontsize=12)
plt.xticks(x, metrics, fontsize=10, rotation=45)
plt.legend(title='Facilities', fontsize=10, loc='upper left')
plt.tight_layout()  # Adjust layout for better readability

# Adding values on top of the bars for better readability
values = [male_counts, female_counts, over_18, under_18, dependants, principals]
for metric_idx in range(len(metrics)):
    for bar_idx, facility_data in enumerate(values[metric_idx]):
        plt.text(x[metric_idx] + (bar_idx - 1) * bar_width, 
                 facility_data + 3, 
                 str(facility_data), 
                 ha='center', va='bottom', fontweight='bold', fontsize=9)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style
sns.set_theme(style="whitegrid")

# Correct data for all metrics for each facility
data = {
    'Facility': ['Complete Care Clinic', 'General Hospital', 'PHC Arogbo'],
    'Male': [219, 288, 19],
    'Female': [160, 195, 10],
    '>18yrs': [235, 330, 17],
    '<18yrs': [144, 153, 12],
    'Dependants': [199, 196, 14],
    'Principal': [180, 287, 15]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Melt the DataFrame for easier grouped bar plotting
df_melted = df.melt(id_vars='Facility', var_name='Metric', value_name='Counts')

# Plotting the grouped bar chart
plt.figure(figsize=(12, 8))
sns.barplot(data=df_melted, x='Metric', y='Counts', hue='Facility', palette='Set2')

# Adding title and labels
plt.title('Comparison of Metrics Across Health Facilities', fontsize=14, fontweight='bold')
plt.xlabel('Metrics', fontsize=12)
plt.ylabel('Counts', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.legend(title='Facility', fontsize=8, loc='upper right')

# Adding values above the bars
for p in plt.gca().patches:
    plt.text(p.get_x() + p.get_width() / 2., 
             p.get_height() + 5, 
             int(p.get_height()), 
             ha='center', va='bottom', fontsize=8, fontweight='bold')

# Show the plot
plt.tight_layout()
plt.show()
```
